```python
%sql
create  table retail_bronze.superstore_bronze.gmart_bronze
  tblproperties (
    delta.columnMapping.mode = 'name'
  )
as
select *, current_timestamp() as ingestion_time
from read_files(
  '/Volumes/retail_bronze/superstore_bronze/raw_superstore/Sample - Superstore.csv',
  format => 'csv',
  header => true,
  inferSchema => true)
```

```python
from pyspark.sql.functions import *

df = (
    spark.read.csv(
        '/Volumes/retail_bronze/superstore_bronze/raw_superstore/Sample - Superstore.csv',
        header=True,
        inferSchema=True
    ).withColumn("Quantity", (col("Quantity").try_cast("integer")))\
        .withColumn("Sales", (col("Sales").try_cast("float")))\
            .withColumn("Discount", round(col("Discount").try_cast("float"),3))\
                .withColumn("Profit", round(col("Profit").try_cast("float")))\
                    .withColumn("IngestionTime", current_timestamp())\
                        .where(col("Quantity")>0)
            )
display(df)
```
