```python
df= spark.table("retail_bronze.superstore_bronze.gmart_bronze")
display(df.count())
```

```python
from pyspark.sql.functions import *
df=df.withColumn("Quantity", (col("Quantity").try_cast("integer")))\
                        .where(col("Quantity") >= 1)\
                            .na.drop(subset=["City","Postal Code"])
```

```python
from pyspark.sql.functions import count, col

df.groupBy(
    "Product ID",
    "Category",
    "Sub-Category",
    "Product Name"
).agg(
    count(col("*")).alias("count")
).select(
    col("Product ID").alias("Product_ID"),
    col("Category").alias("Category"),
    col("Sub-Category").alias("Sub-Category"),
    col("Product Name").alias("Product_Name")
).write.saveAsTable(
    "retail_silver.superstore_silver.product_silver"
)
```
